# Median House Prices By Neighbourhood

```python
# Group the data to calculate median house prices by neighborhood
median_prices = df.groupby('Neighborhood')['House_Prices_Median'].median().reset_index()
```

```python
# Bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Neighborhood', y='House_Prices_Median', data=median_prices, palette='viridis')
plt.title('Median House Prices by Neighborhood', fontsize=14)
plt.xlabel('Neighborhood')
plt.ylabel('Median House Price ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

The house prices appear relatively close across neighborhoods, suggesting a narrower range in median values than expected.
Staten Island and Manhattan typically have higher median house prices.
Bronx and Brooklyn show slightly lower median house prices compared to Staten Island or Queens, which aligns with their historical market trends.
